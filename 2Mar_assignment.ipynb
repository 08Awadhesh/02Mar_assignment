**Q1:** What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.




**Answer**: Matplotlib is a popular data visualization library in Python. It provides a wide range of tools for creating static, animated, and interactive visualizations in various formats. Matplotlib is widely used in scientific computing, data analysis, and other fields where data visualization is essential.

Here are five plots that can be created using the Pyplot module of Matplotlib:

**(I) Line Plot:** This is a basic plot where data points are connected by straight lines. It is useful for visualizing trends and patterns in data over time or any continuous variable.

**(II) Bar Plot:** Bar plots represent categorical data using rectangular bars. They are useful for comparing values between different categories or groups.

**(III)Scatter Plot:** Scatter plots display individual data points as dots on a two-dimensional coordinate system. They are used to identify relationships and patterns between two variables.

**(IV)Histogram:** Histograms represent the distribution of a dataset by dividing it into bins and displaying the frequency or proportion of data points within each bin. They are useful for visualizing the underlying distribution of numerical data.

**(V) Pie Chart:** Pie charts display data as a circular graph, where the size of each slice represents the proportion of each category or value in the dataset. They are useful for illustrating the composition or relative contributions of different parts to a whole.

**Q3:** Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

**Answer**:The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to arrange multiple plots in a grid-like structure, where each plot represents a different aspect or subset of your data.

There are several reasons why the subplot() function is useful:

Multiple Plots in a Single Figure: With subplot(), you can create a single figure and divide it into multiple subplots, each displaying different data or different views of the same data. This is particularly helpful when you want to compare multiple plots side by side or present related information in a compact and organized manner.

Layout Flexibility: The subplot() function provides flexibility in defining the layout of subplots. You can specify the number of rows and columns in the grid, as well as the position of each subplot. This allows you to create different arrangements, such as a grid of equally sized subplots or a mix of differently sized subplots.

Easy Navigation and Indexing: Each subplot created using subplot() is assigned a unique index, making it easy to reference and modify specific subplots within the grid. This indexing scheme simplifies tasks such as modifying plot properties, adding annotations, or updating the content of individual subplots.

Consistent Figure Size: When using subplot(), the overall size of the figure is determined by the number of rows, columns, and the size of each subplot. This ensures that all subplots within the figure have a consistent size and aspect ratio, resulting in a visually cohesive presentation.

Simplified Plotting Workflow: By using the subplot() function, you can streamline your plotting workflow by organizing related plots within a single figure. This helps maintain the context and avoids the hassle of managing multiple separate figure windows

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```

**Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])


**Answer**: A bar plot is a type of chart that represents categorical data using rectangular bars. Each bar typically corresponds to a category or group, and the length or height of the bar represents a specific value associated with that category. Bar plots are commonly used to compare and display the quantitative values of different categories or to show the distribution of a single categorical variable.

Bar plots are useful for the following reasons:

**Comparison:** Bar plots allow easy visual comparison between different categories or groups. By visually comparing the lengths or heights of the bars, you can quickly identify the relative differences in the values being represented.

**Categorical Data Visualization:** Bar plots are effective in displaying categorical data, where each bar represents a discrete category. They provide a clear representation of the distribution of data and help in identifying patterns, trends, or outliers within each category.

**Simple and Intuitive:** Bar plots are simple to understand and interpret. The length or height of the bars directly corresponds to the magnitude of the value being represented, making it easy to grasp the information presented in the plot

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("comany with their profit")
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

**Answer**: A box plot, also known as a box-and-whisker plot, is a statistical chart used to display the distribution of a continuous numerical variable or a set of variables. It provides a concise summary of key statistical measures and the shape of the dataset, including the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values.

Box plots are useful for the following reasons:

**Distribution Visualization:** Box plots offer a visual representation of the distribution of data. The central box represents the interquartile range (IQR), which contains the middle 50% of the data. The line inside the box represents the median value. The whiskers extend from the box to the minimum and maximum values, excluding outliers.

**Identification of Outliers:** Box plots provide information about potential outliers in the dataset. Points lying outside the whiskers are considered outliers and can be easily identified, helping to identify data points that deviate significantly from the main distribution.

**Comparison of Distributions:** Box plots are effective for comparing the distributions of different groups or categories. Multiple box plots can be displayed side by side, allowing for easy visual comparison of key statistical measures between groups and detecting differences in the shape, spread, or skewness of the data.

**Detection of Skewness:** Box plots can reveal the presence of skewness in the dataset. Skewed distributions tend to have asymmetrical box plots, indicating a longer tail on one side of the median.

**Data Summary:** Box plots summarize the key statistical measures of a dataset in a compact and intuitive manner. These measures include the minimum and maximum values, quartiles, and median, providing an overview of the dataset's central tendency, spread, and overall shape.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)


data = [box1, box2]


fig, ax = plt.subplots(figsize=(8, 6))


ax.boxplot(data)


ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box Plot')


plt.show()
```
